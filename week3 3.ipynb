```python
import urllib.request
```

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
page=urllib.request.urlopen(url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
all_tables=soup.find_all("table")
```

```python
right_table=soup.find('table', class_='wikitable sortable')
```

```python
A=[]
B=[]
C=[]
for row in right_table.findAll("tr"):
    cells=row.findAll("td")
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
def remove_useless(lst):
    lst_r=[]
    for i in lst:
        i=i[:-1]
        lst_r.append(i)
    return lst_r
        
```

```python
A=remove_useless(A)
B=remove_useless(B)
C=remove_useless(C)
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=["PostalCode"])
df["Borough"]=B
df["Neigborhood"]=C
df
```

```python
df_valid=df[df["Borough"]!="Not assigned"]
df_valid=df_valid.reset_index().drop(["index"],axis=1)
df_valid
```

```python
for i in range(len(df_valid)):
    if df_valid.loc[i,"Neigborhood"]=="Not assigned":
        df_valid.loc[i,"Neigborhood"]=df_valid.loc[i,"Borough"]
```

```python
df_valid.shape
```

```python
geo_data=pd.read_csv("E:/Coursea/capstone/geo.csv")
geo_data.head()
```

```python
geo=pd.DataFrame(columns=geo_data.columns)
for i in A:
    part=geo_data[geo_data["Postal Code"]==i]
    geo=pd.concat([geo,part])
print(geo)
```

```python
lat=geo["Latitude"].values.tolist()
long=geo["Longitude"].values.tolist()
```

```python
df_valid["Latitude"]=lat
df_valid["Longitude"]=long
df_valid
```

```python
df_valid.shape
```

```python
toronto_df=df_valid[df_valid["Borough"].str.contains("Toronto")]
toronto_df.head()
```

```python
toronto_df["Borough"].unique()
```

```python
from sklearn.cluster import KMeans
```

```python
k=4
cluster_df=toronto_df[["Latitude","Longitude"]]
kmeans=KMeans(n_clusters=k,random_state=0).fit(cluster_df)
toronto_df["Cluster Label"]=kmeans.labels_
toronto_df
```

```python
from geopy.geocoders import Nominatim
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="na")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
import folium
```

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Neigborhood'], toronto_df['Cluster Label']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
